# Rodando a sua IA no Google Colab

Aqui está importando as bibliotecas necessarias e conectando com a chave APi GEMINI

```python
import os
from google.colab import userdata
# Configuração da Chave
os.environ["GOOGLE_API_KEY"] = userdata.get('GEMINI_API_KEY')
```

```python
# Conexão com SDK
from google import genai
# Criamos com o client
client = genai.Client()
```

```python
# Chamada do modelo() junto com o conteudo que vc deseja que ele retorne
resposta = client.models.generate_content(model="gemini-2.5-flash", contents="O que é a inteligência Artifical ?")
```

```python
# Para exibir o texto da resposta no console
print(resposta.text)
```

Vantagem de ter um chatBo e da sua memória. O chat consegue armazenar as perguntas e respostas anteriores e as envias automaticamente junto com a nova pergunta, permitindo que você mantenha uma nova conversa fluida.

```python
# Criação de um chatBot
chat = client.chats.create(model ="gemini-2.5-flash")
```

```python
# Enviar uma mensagem
resposta = chat.send_message("O que é a inteligência Artifical? ")
```

```python
print(resposta.text)
```

```python
resposta = chat.send_message("Quando foi lançado o ChatGPT")
print(resposta.text)
```

```python
# get_history serve para exibir o histórico do ChatBot
chat.get_history()
```

```python
# Criando um loop onde o usuário consegue fazer várias perguntas para ChatBot, quando não quiser continuar mais ele digite "fim" para encerrar.

prompt = input("Digite a sua pergunta: ")

while prompt != "fim":
  resposta = chat.send_message(prompt)
  print(resposta.text)
  print("\n")
  prompt = input("Digite a sua pergunta: ")
```

Desafio: Crie um código que percorra a lista de médias e some 1 a cada valor, exceto quando o valor passar de 10.0, que aí ele deve parar em 10.0

```python
n = 0

while n < len(lista_de_medias):

  lista_de_medias[n] = lista_de_medias[n] + 1.0

  if lista_de_medias[n] > 10.0:
    lista_de_medias[n] = 10.0

  n = n + 1

print(lista_de_medias)

```
